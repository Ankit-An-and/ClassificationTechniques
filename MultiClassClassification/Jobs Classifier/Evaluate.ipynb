```python
import numpy as np
import joblib
from ipynb.fs.full.Common import evaluate
```

```python
# load models

model_dectree = joblib.load('jobs_dectree.joblib')
model_knn = joblib.load('jobs_knn.joblib')
model_rf = joblib.load('jobs_rf.joblib')
model_naivebayes = joblib.load('jobs_naivebayes.joblib')
model_mlp = joblib.load('jobs_mlp.joblib')
```

```python
# Load testing data

loaded = np.load('jobs_data.npz', allow_pickle=True)
X_test = loaded['Xtest']
y_test = loaded['ytest']
```

```python
# Evaluation Decision Tree

pred_dectree = model_dectree.predict(X_test)
evaluate(y_test, pred_dectree, "Decision Tree")
```

```python
# Evaluation KNN

pred_knn = model_knn.predict(X_test)
evaluate(y_test, pred_knn, 'K Nearest Neighbours')
```

```python
# Evaluation Random Forest

pred_rf = model_rf.predict(X_test)
evaluate(y_test, pred_rf, 'Random Forest')
```

```python
# Evaluation Naive Bayes

pred_nb = model_naivebayes.predict(X_test)
evaluate(y_test, pred_nb, 'Naive Bayes')
```

```python
# Evaluation MLP Neural Network

pred_mlp = model_mlp.predict(X_test)
evaluate(y_test, pred_mlp, 'MLP Neural Network')
```
