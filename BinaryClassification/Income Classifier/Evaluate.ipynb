```python
import numpy as np
import joblib
from ipynb.fs.full.Common import evaluate
```

```python
# load models

model_dectree = joblib.load('adult_income_decisiontree.joblib')
model_knn = joblib.load('adult_income_knn.joblib')
model_logreg = joblib.load('adult_income_logreg.joblib')
model_naivebayes = joblib.load('adult_income_naivebayes.joblib')
model_svm = joblib.load('adult_income_svm.joblib')
model_mlp = joblib.load('adult_income_mlp.joblib')
```

```python
# Load testing data

loaded = np.load('adult_income_data.npz', allow_pickle=True)
X_test = loaded['Xtest']
y_test = loaded['ytest']
```

```python
# Evaluation Decision Tree

pred_dectree = model_dectree.predict(X_test)
evaluate(y_test, pred_dectree, "Decision Tree")
```

```python
# Evaluation KNN

pred_knn = model_knn.predict(X_test)
evaluate(y_test, pred_knn, 'K Nearest Neighbours')
```

```python
# Evaluation logistic Regression

pred_logreg = model_logreg.predict(X_test)
evaluate(y_test, pred_logreg, 'Logistic Regression')
```

```python
# Evaluation Naive Bayes

pred_nb = model_naivebayes.predict(X_test)
evaluate(y_test, pred_nb, 'Naive Bayes')
```

```python
# Evaluation SVM

pred_svm = model_svm.predict(X_test)
evaluate(y_test, pred_svm, 'Support Vector Machine')
```

```python
# Evaluation MLP Neural Network

pred_mlp = model_mlp.predict(X_test)
evaluate(y_test, pred_mlp, 'MLP Neural Network')
```
